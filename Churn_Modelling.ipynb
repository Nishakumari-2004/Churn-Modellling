```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path="/content/drive/MyDrive/Colab Notebooks/Churn_Modelling.csv"
dataset=pd.read_csv(path)
dataset.head()
```

```python
dataset.shape
```

```python
X=dataset.iloc[:,3:13]    #independent features
y=dataset.iloc[:,13]
```

```python
# Create dummy variables since these are categorical variables,convert into numerical
geography=pd.get_dummies(X["Geography"],drop_first=True)
gender=pd.get_dummies(X["Gender"],drop_first=True)
```

```python
X=pd.concat([X,geography,gender],axis=1)
X
```

```python
 X=X.drop(['Geography','Gender'],axis=1)
X
```

```python
#Splitting the dataset into training and testing set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense,InputLayer,Activation,Embedding,Flatten,BatchNormalization
from scikeras.wrappers import KerasClassifier
from keras.activations import relu,sigmoid
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
```

```python
#Initialising the ANN
classifier=Sequential()
```

```python
# Add the input layer
classifier.add(InputLayer(shape=(11,)))
```

```python
# adding the first hidden layer
classifier.add(Dense(units=10,kernel_initializer='he_uniform',activation='relu'))
# relu works well in hidden layers ,it prevents vanishing gradient problem
#can use he normal also apart from he uniform ,that also works well with relu
```

```python
#adding the second hidden layer
classifier.add(Dense(units=20,kernel_initializer='he_uniform',activation='relu'))
```

```python
#adding the third hidden layer
classifier.add(Dense(units=15 ,kernel_initializer='he_uniform',activation='relu'))
```

```python
#adding the output layer
classifier.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
#Compiling  the ANN ,this is binary classification
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Fitting ANN to the training set
model_history=classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=100)
```

```python
#Predicting the Test set results
y_pred=classifier.predict(X_test)
y_pred=(y_pred>0.5)
```

```python
# making the confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
#accuracy score
from sklearn.metrics import accuracy_score
score= accuracy_score(y_pred,y_test)
score
```
